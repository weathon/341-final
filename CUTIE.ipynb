```python
from openai import OpenAI
from google.colab import userdata
import requests
import pylab


client = OpenAI(api_key=userdata.get('key'))
last_prompts = []
for i in range(10):
  response = client.images.generate(
    model="dall-e-3",
    prompt="A very cheesy cheese buger, photo style. Your last prompt were ' " +str(last_prompts)+" ' but they were not cheesy enough. I want much cheesier than those. ",
    size="1024x1024",
    quality="standard",
    n=1,
  )

  image_url = response.data[0].url
  print(image_url)
  last_prompts.append(response.data[0].revised_prompt)

  img_data = requests.get(image_url).content
  with open(f'image{i}.jpg', 'wb') as handler:
    handler.write(img_data)

  pylab.imshow(pylab.imread(f'image{i}.jpg'))
  pylab.show()
```

```python
from openai import OpenAI
from google.colab import userdata
import requests
import pylab


client = OpenAI(api_key=userdata.get('key'))
last_prompts = []
for i in range(10):
  response = client.images.generate(
    model="dall-e-2",
    prompt="A very cute cat eating a cheesebuger, photo style. Your last prompt were ' " +str(last_prompts)+" ' but they were not cute enough. I want much cuter than those. ",
    size="1024x1024",
    quality="standard",
    n=1,
  )

  image_url = response.data[0].url
  print(image_url)
  last_prompts.append(response.data[0].revised_prompt)

  img_data = requests.get(image_url).content
  with open(f'image{i}.jpg', 'wb') as handler:
    handler.write(img_data)

  pylab.imshow(pylab.imread(f'image{i}.jpg'))
  pylab.show()
```
